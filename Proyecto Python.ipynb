```python
#variables

def linealRegression(y, X, epochs, print_error, lr):
    #Constantes 
    n = 0
    m = 1
    b = 1   
    
    #Inicializacion de vectores
    vector_errores = []
    vector_mb = np.empty((epochs,2))
    vector_yhat = np.empty((epochs,y.shape[1]))
    matX = (np.vstack((X,np.ones(y.shape[1])))).T
    mb = np.array([m,b])
 
    #Ciclo para realizar el modelo de regresion lineal.
    for i in range(0,epochs):  
        yhat = np.dot(matX,mb)
        error = 1/2*(np.mean((y-yhat)**2))
        variacion = (np.array([yhat-y,yhat-y])).T
        grad = np.mean(variacion*matX)
        mb = mb - lr*grad
        vector_errores.append(error)
        vector_mb[i] = mb
        vector_yhat[i] = yhat
        n += 1

        if(n == print_error):
            n = 0
            print(error)
            
    return(np.array([yhat,vector_errores, vector_mb, vector_yhat, matX])) 

    

ynew, errores, vect_mb, vect_yhat, matX = linealRegression(Ytr0, Tra1, 200, 20, 0.001)
ynew1, errores1, vect_mb1, vect_yhat1, matX1 = linealRegression(Ytr0, Tra2, 100, 50, 0.0000000000002)
```

```python
def graphError(error, epochs):
    xr = np.array(np.linspace(0,epochs,epochs))
    yr = np.array(error)

    fig = plt.figure(figsize=(14,8))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title('Error')
    plt.plot(xr,yr)
    return plt.show()
    
graphError(errores, 200)
graphError(errores1, 100)
```

```python

def graficarModelo(x,y,vect_mb, matX, n):
    fig = plt.figure(figsize=(14,8))
    plt.scatter(x, y)
    plt.title("Estimacion del precio de las viviendas")
    plt.ylabel("Precio de venta de casas")
    
    d = vect_mb.shape[0] + 1
    for i in range(0, d, n):
        yhat = np.dot(matX,vect_mb[i])
        plt.plot(np.squeeze(np.asarray(x)), yhat)
    return plt.show()


graficarModelo(Tra1, Ytr0, vect_mb, matX,  15)
graficarModelo(Tra2, Ytr0, vect_mb1, matX1,  15)
```
